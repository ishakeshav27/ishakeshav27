```python
df.info()

columns_name = df.columns
null_data = df.isnull().sum()
not_null_data = len(df) - null_data

plt.figure(figsize=(15, 5))
plt.bar(columns_name, null_data, color="#ff746c", alpha = 0.6)
plt.bar(columns_name, not_null_data, bottom = null_data, color = "#08bcc4", alpha = 0.6)
plt.legend(('Null', 'Not null'), loc='best')
plt.show()
```

```python
pclass_survived = df[df['Survived'] == 1]['Pclass']
pclass_not_survived = df[df['Survived'] ==0]['Pclass']

x_class = ['1st Class', '2nd Class', '3rd Class']
y_class_survived = pclass_survived.value_counts().sort_index()
y_class_not_survived = pclass_not_survived.value_counts().sort_index()

plt.bar(x_class, y_class_survived, color = "#08bcc4", alpha = 0.6)
plt.bar(x_class, y_class_not_survived, bottom = y_class_survived, color = "#7d3227", alpha = 0.6)
plt.legend(('Survived', 'Not Survived'), loc='best')
plt.title('Survivors by Class Group')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# survivors by gender
sex_survived = df[df['Survived'] == 1]['Sex']
sex_not_survived = df[df['Survived'] == 0]['Sex']

x_sex = ['female', 'male']
y_sex_survived = sex_survived.value_counts().sort_index()
y_sex_not_survived = sex_not_survived.value_counts().sort_index()

plt.bar(x_sex, y_sex_survived, color = "#08bcc4", alpha = 0.6)
plt.bar(x_sex, y_sex_not_survived, bottom = y_sex_survived, color = "#183d59", alpha = 0.6)
plt.legend(('Survived', 'Not Survived'), loc='best')
plt.title('Survivors by Gender')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
#Survivors Age Group
df_train=df
age_survived = df_train[df_train['Survived'] == 1]['Age']
age_not_survived = df_train[df_train['Survived'] == 0]['Age']

plt.hist([age_survived, age_not_survived], bins=5, range=(1, max(df_train['Age'])), stacked=True, color = ["#12030b", "#750f41"], alpha = 0.6)
plt.legend(('Survived', 'Not survived'), loc='best')
plt.title('Survivors by Age Group')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#Feature Engineering
df_train = df_train[['Pclass', 'Sex', 'Age', 'Survived']]

# Pclass
c = ['1st class', '2nd class', '3rd class']

for i in range(3) :

  df_train.loc[df_train['Pclass'] == i+1, c[i]] = 1
  df_train.loc[df_train['Pclass'] != i+1, c[i]] = 0
    
#gender
df_train.loc[df_train['Sex'] == 'male', 'Gender'] = 1
df_train.loc[df_train['Sex'] == 'female', 'Gender'] = 0

# age
df_train['Age'] = pd.cut(df_train['Age'], bins=[1, 15, 25, 56, np.inf], labels=[1, 2, 3, 4])
groupAge = ['kids', 'teen', 'adult', 'elderly']

for i in range(4):

  df_train.loc[df_train['Age'] == i+1, groupAge[i]] = 1
  df_train.loc[df_train['Age'] != i+1, groupAge[i]] = 0
    
df_train.drop(['Pclass', 'Sex', 'Age'], axis=1, inplace=True)

df_train.head(10)
```
